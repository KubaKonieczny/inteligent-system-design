## Changing real match stats for the average statistics for each team

We believe that this is a better approach than learning the model on the real match statistics. If the model would be trained on the real data it would probably learn the team with a higher number of goals is the winner, what would lead to lower quality of predictions.

```python
teams_df.head()
```

As a test set we were using matches provided in task description. However, we decided to exclude 2 matches of Cape Verde, because they weren't actually played, as they were just walkovers. If we would include them in our test set it would lead to false results as the outcome is not based on skill of the players (which can be represented by stistics) but some other factors.

It is worth meantioning that we used all matches for creating average for each team as the data set is small. If we were to exclude those matches, some teams would only have few matches taken into accunont.
